# Bar Plot

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
subjects = ['Maths', 'English', 'Science', 'Social Studies', 'Computer']
Marks = [89, 90, 45, 78, 99]
```

```python
plt.bar(subjects, Marks)
plt.show()
```

```python
plt.bar(subjects, Marks, color = 'red')
plt.show()
```

```python
colors = ['red', 'blue', 'green', 'orange', 'purple']
plt.bar(subjects, Marks, color = colors, alpha =0.7, width=0.6, edgecolor = 'black', linewidth = 2, linestyle = '-.')
plt.show()
```

```python
plt.barh(subjects, Marks)
plt.show()
```

Multiple plot in same figure

```python
subjects = ['Maths', 'English', 'Science', 'Social Studies', 'Computer']
Marks1 = [89, 90, 45, 78, 99]
Marks2 = [78, 56, 34, 90, 50]
```

```python
plt.figure(figsize=(6,6))

plt.bar(subjects, Marks1, )
plt.bar(subjects, Marks2, color= colors, alpha=0.5)

plt.show()
```

```python
plt.figure(figsize=(6,6))
sub_len = np.arange(len(subjects))
width = 0.4

plt.bar(sub_len, Marks1, width = width, color= colors)
plt.bar(sub_len + width, Marks2, color= colors, width = width, alpha=0.5)

plt.show()
```

# Histogram

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
marks_50_students = np.random.randint(0,100, (50))
marks_50_students
```

```python
bins = np.arange(0,100,5)
plt.hist(marks_50_students, bins=bins)
plt.xticks(np.arange(0,100,5))
plt.show()
```

```python
marks1 = np.random.randint(1,100,50)
marks2 = np.random.randint(0,100,50)
```

```python
bins = np.arange(0,100,5)

plt.figure(figsize=(6,6))
plt.hist([marks1, marks2], bins=bins, color =['blue', 'orange'])
plt.xticks(np.arange(0,100,5))
plt.show()
```

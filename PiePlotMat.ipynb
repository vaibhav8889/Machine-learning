# Pie Plot

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
classess = ['Physics', 'Chemistry', 'Maths', 'Science', 'SST']
marks = [89, 45, 78, 23, 90]
```

```python
plt.pie(marks, labels=classess, autopct='%1.1f%%')
plt.show()
```

```python
explode_values = [0.1, 0, 0, 0, 0]

plt.pie(marks, labels=classess, autopct='%1.1f%%', explode=explode_values, shadow = True)
plt.show()
```

```python
explode_values = [0.1, 0, 0, 0, 0]
textprops = {'fontsize': 15, 'color':'w'}
wedgeprops = {'linewidth': 0.5, 'linestyle':'-', 'edgecolor':'black'}
plt.pie(marks, labels=classess, autopct='%1.1f%%', explode=explode_values, shadow = True, textprops=textprops, wedgeprops=wedgeprops)
plt.legend()
plt.show()
```
